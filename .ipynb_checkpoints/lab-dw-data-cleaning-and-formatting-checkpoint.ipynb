## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
customer_data['gender'].unique()  # check the unique elements in the column
```

```python
# I will define a function to clean the gender values

def clean_gender(x):
    if x in ['M', 'Male']:
        return 'M'
    elif x in ['F','Femal','female']:
        return 'F'
    else:
        return 'U'

customer_data['gender'] = list(map(clean_gender, customer_data['gender']))
customer_data['gender'].unique()  # To check the results again
```

```python
print(customer_data['state'].unique())  # check the unique elements in the column
print(customer_data.dtypes)
```

```python
# I will define a function to clean the state values

def clean_state(x):
    if x in ['Washington', 'WA']:
        return 'Washington'
    elif x in ['Arizona','AZ']:
        return 'Arizona'
    elif x in ['California','Cali']:
        return 'California'
    elif x in ['Nevada']:
        return 'Nevada'
    elif x in ['Oregon']:
        return 'Oregon'
    else:
        return 'U'
        

customer_data['state'] = list(map(clean_state, customer_data['state']))
customer_data['state'].unique()  # To check the results again
```

```python
customer_data['education'].unique()  # check the unique elements in the column
```

```python
# I replace 'Bachelors' values to 'Bachelor' to improve consistency

customer_data['education'].replace('Bachelors', 'Bachelor', inplace=True)
customer_data['education'].unique()  # check if its okay
```

```python
# I replace the % character in the whole column and make it numeric

customer_data['customer_lifetime_value'] = customer_data['customer_lifetime_value'].str.replace('%', '')

customer_data['customer_lifetime_value'] =  pd.to_numeric(customer_data['customer_lifetime_value'], errors='coerce' )

print(customer_data.dtypes)
```

```python
# I divide the whole column by 100 in order to convert it finally into a float

customer_data['customer_lifetime_value'] = customer_data['customer_lifetime_value']/100
customer_data
```

```python
customer_data.dtypes

# I change the data type of the column

customer_data['customer_lifetime_value'] =  pd.to_numeric(customer_data['customer_lifetime_value'], errors='coerce' )
customer_data

# Lets check if its okay
customer_data.dtypes
```

```python
print(customer_data['vehicle_class'].unique())  # check the unique elements in the column
print(customer_data.dtypes)
```

```python
# In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" I replaced by "Luxury"

customer_data['vehicle_class'] = customer_data['vehicle_class'].str.replace('Luxury Car', 'Luxury')
customer_data['vehicle_class'] = customer_data['vehicle_class'].str.replace('Luxury SUV', 'Luxury')
customer_data['vehicle_class'] = customer_data['vehicle_class'].str.replace('Sports Car', 'Luxury')
customer_data

print(customer_data['vehicle_class'].unique())  # check if its okay
customer_data
```

```python
# X-y split

y = customer_data['total_claim_amount']
X = customer_data.drop(['total_claim_amount'], axis=1)

X_num = X.select_dtypes(include = np.number)
X_cat = X.select_dtypes(include = object) 
display(X_num)
X_cat
```

```python
numerical.columns
```

```python
# Scaling
# I will do it with Normalizing

# Normalizing data: make data range from 0 - 1, instead of from min to max
transformer = MinMaxScaler().fit(X_num) #.fit takes only min and max
x_normalized = transformer.transform(X_num)
print(x_normalized.shape)
x_normalized
normalized_num = pd.DataFrame(x_normalized, columns=X_num.columns)
normalized_num
```

Round 6
- One Hot/Label Encoding (categorical).
- Concat DataFrames

```python
X_cat
```

```python
# One Hot/Label Encoding (categorical)

# one hot encoding is a way to turn categorical variables into multiple numerical columns
from sklearn.preprocessing import OneHotEncoder

#I exclude custumer column as I dont want it to train my model
X_cat_without_customer = X_cat.drop('customer', axis=1)


encoder = OneHotEncoder(drop='first').fit(X_cat_without_customer)  # This is useful for put our data into the model. The less variables the better so we drop a redundant one.
print(encoder.categories_)
encoded = encoder.transform(X_cat_without_customer).toarray()

#this makes the new columns with the proper name
cols = encoder.get_feature_names_out(input_features=X_cat_without_customer.columns)

print(encoded)
onehot_encoded = pd.DataFrame(encoded, columns=cols)
onehot_encoded
onehot_encoded.head(20)
```

```python
X_scaled_encoded = pd.concat([normalized_num, onehot_encoded], axis=1)
X_scaled_encoded.head()
```

Linear Regression
- Train-test split.
- Apply linear regression.

```python
# Train-test split

# defining our y
y = customer_data['total_claim_amount']
y.head()
```

```python
X_scaled_encoded.shape
X = X_scaled_encoded
```

```python
# train test split is the way ML generates its claim to fame: 
# we build the model on a portion of the data but we then validate it in another "fresh" portion.
# Our model has no opportunity to "cheat": it must accurately guess the values 
# in the "fresh" dataset that it never saw before
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train.head()
```

```python
y_train.head()

# it has the same indexes as the X_train so it's done perfectly
```

```python
# We train/fit our model

lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
```

Model Validation
Description:
- R2.
- MSE.
- RMSE.
- MAE.

```python
from sklearn.metrics import r2_score, mean_absolute_error
predictions = lm.predict(X_train)
r2_score(y_train, predictions)
```

```python
# But now we evaluate it in the TEST portion of the data, that we did not use for training.
# This way we know our model is genuinely guessing our donations, not just repeating the values it has seen in the training data

predictions_test = lm.predict(X_test)
r2_score(y_test, predictions_test)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,predictions_test)
mse
```

```python
rmse = np.sqrt(mean_squared_error(y_test,predictions_test))
rmse

# This is the margin of error we should use as it is in the right format as the original variable total_claim_amount
# rmse is USD and mse is in squared USD.
```

```python
mae = mean_absolute_error(y_test, predictions_test)
print(mae)
```
