```python
import pandas as pd
import numpy as np

# tryout dataframe

customer_data = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
customer_data
```

```python
customer_data['Education'].unique()
```

```python
## FIRST FUNCTION

def replace_str_values(df, c, x, y):

    return df[c].str.replace(x, y)
```

```python
customer_data['Education'] = replace_str_values(customer_data, 'Education', 'Bachelors', 'Bachelor') 
```

```python
customer_data['Education'].unique()
```

```python
## SECOND FUNCTION

def rename_col(df, x, y):
    return df.rename(columns = {x:y}, inplace = True) 
```

```python
rename_col(customer_data, 'ST', 'state')
```
